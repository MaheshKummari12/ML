```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
#Using the dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('dendogram')
plt.xlabel('customer')
plt.ylabel('Eucledian distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage = 'ward')
y_hc = ac.fit_predict(X)
```

```python
np.unique(y_hc)
```

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],c='r',s=100,label='Cluster1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],c='g',s=100,label='Cluster2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],c='y',s=100,label='Cluster3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],c='magenta',s=100,label='Cluster4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],c='b',s=100,label='Cluster5')

plt.title('Herirarchical clustering')
plt.xlabel('nnual Income (k$)')
plt.ylabel('spending score')
plt.legend()
plt.show()

```
