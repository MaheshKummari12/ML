```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Mall_customers.csv')
data

```

```python
X = data.iloc[:,[3,4]].values
```

```python
X.shape
```

Using the elbow method finding the optimal no of clusters


```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=4)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('K-means clustering')
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

training the K-Means on the whole dataset

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=4)
y_means = kmeans.fit_predict(X)
```

```python
np.unique(y_means)
```

Visulaizing the clusters

```python
plt.scatter(X[y_means==0,0], X[y_means==0,1],s = 100,c='r',label='Cluster1')
plt.scatter(X[y_means==1,0], X[y_means==1,1],s = 100,c='b',label='Cluster2')
plt.scatter(X[y_means==2,0], X[y_means==2,1],s = 100,c='g',label='Cluster3')
plt.scatter(X[y_means==3,0], X[y_means==3,1],s = 100,c='magenta',label='Cluster4')
plt.scatter(X[y_means==4,0], X[y_means==4,1],s = 100,c='y',label='Cluster5')

plt.title('K-means for clustering')
plt.xlabel('no of clusters')
plt.ylabel('WCSS')
plt.legend()
plt.show()
```
