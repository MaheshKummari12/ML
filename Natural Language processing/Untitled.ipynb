```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
text = ['call you tonight', 'Call me a cab', 'Please call me... PLEASE!']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit(text)
```

```python
cv.get_feature_names_out()
```

```python
X = cv.transform(text)
```

```python
print(X)
```

```python
X.toarray()
```

```python
x = pd.DataFrame(X.toarray(), columns=cv.get_feature_names_out())
```

```python
print(x)
```

```python
simple_test= ["please mahesh don't call me"]
```

```python
y = cv.transform(simple_test)
y.toarray()
```

```python
print(y)
```

```python
y = pd.DataFrame(y.toarray(), columns=cv.get_feature_names_out())
```
