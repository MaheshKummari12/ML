```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset.head(5)
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

Training the whole dataset on Linear model


```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X,y)
```

```python
regression.score(X,y)
```

Training the whole dataset on polynomial function


```python
from sklearn.preprocessing import PolynomialFeatures
PL = PolynomialFeatures(degree = 6)
X_poly = PL.fit_transform(X)
X_regression_2 = LinearRegression()
X_regression_2.fit(X_poly,y)
```

```python
X_regression_2.score(X_poly,y)
```

Visualizing the data for linear regression model


```python
plt.scatter(X,y,c='b')
plt.plot(X,regression.predict(X),c='r')
plt.title='Linear regression model'
plt.xlabel='Level'
plt.ylabel='Salary'
plt.show()
```

Visualzing the data for polynomial model

```python
plt.scatter(X,y,c='b')
plt.plot(X,X_regression_2.predict(PL.fit_transform(X)),c='r')
plt.title='Polynomial regression model'
plt.xlabel='Level'
plt.ylabel='Salary'
plt.show()
```

Visualizing the polynomial regression results with higher resolution and smoother curve

```python
X_grid = np.arange(min(X.ravel()),max(X.ravel()),0.1)
#Converting a 2D array into 1D array
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,c='r')
plt.plot(X_grid,X_regression_2.predict(PL.fit_transform(X_grid)),c='b')
plt.title='Polynomial regression model'
plt.xlabel='Level'
plt.ylabel='Salary'
plt.show()
```
